# **Data in Motion Weekly pandas challenge 3**


##Challenge Questions

1.    Which continent drinks more beer on average?
2.    For each continent print the statistics for wine consumption.
3.    Print the mean alcohol consumption per continent for every column.
4.    Print the median alcohol consumption per continent for every column.

## **Preparation**

```python
#Dependencies
import pandas as pd
import numpy as np
```

```python
#importing data
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'

data = pd.read_csv(url)

data.head()
```

```python
#rows and columns
data.shape
```

```python
#There are some NaN values in the continent column
data.isnull().sum()
```

```python
#Displaying the rows where continent is NaN
data[data.isna().any(axis=1)]
```

```python
df = data.copy()
```

```python
#The missing continent values are all in North America 
df['continent'].fillna('NA', inplace=True)
```

```python
#Now there are no missing values
df.isnull().sum()
```

## **1. Which continent drinks more beer on average?**

```python
#Grouping results by continent and displaying the mean values in descending order.
data.groupby('continent')['beer_servings'].mean().sort_values(ascending=False)
```

### **Answer:** Europe drunks the most servings of beer on average

## 2. **For each continent print the statistics for wine consumption.**

```python
#Europe also drinks the most wine on average.
df.groupby('continent')['wine_servings'].describe().sort_values('mean', ascending=False)
```

## **3. Print the mean alcohol consumption per continent for every column.**

```python
df.pivot_table(index='continent', aggfunc='mean')
```

## **4. Print the median alcohol consumption per continent for every column.**

```python
df.pivot_table(index='continent', aggfunc='median')
```
