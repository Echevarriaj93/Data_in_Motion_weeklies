```python
import pandas as pd
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data' 

wine = pd.read_csv(url) 
```

## 1.    Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns.

```python
wine.drop(wine.columns[[0,3,6,8,10,12,13]], axis = 1, inplace= True)
```

## 2.    Assign the columns as below:
    - alcohol
    - malic_acid
    - alcalinity_of_ash
    - magnesium
    - flavanoids
    - proanthocyanins
    - hue

```python
wine.columns = ['alcohol','malic_acid','alcalinity_of_ash','magnesium',' flavanoids', 'proanthocyanins', 'hue']
```

```python
wine.head()
```

## 3.    Set the values of the first 3 rows in the alcohol column as NaN

```python
wine['alcohol'] = np.where(wine['alcohol'].isin(wine['alcohol'][0:3]), np.nan, wine['alcohol'])
```

```python
wine.head()
```

## 4.    Now set the value of the rows 3 and 4 of the magnesium column as NaN

```python
wine['magnesium'] = np.where(wine['magnesium'].isin(wine['magnesium'][2:4]), np.nan, wine['magnesium'])
```

```python
wine.head()
```

## 5.    Fill in the null values (NaN) with the number 10 in the alcohol column and 100 in magnesium column.

```python
wine['alcohol'] = wine['alcohol'].replace(np.nan, 10)
```

```python
wine['alcohol'].head()
```

```python
wine['magnesium'] = wine['magnesium'].replace(np.nan, 100)
```

```python
wine['magnesium'].head()
```

## 6.    Count the number of missing values in the entire dataset.

```python
wine.isnull().sum()
```
