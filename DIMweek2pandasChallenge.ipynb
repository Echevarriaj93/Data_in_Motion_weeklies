## **Cleaning the Data**

```python
#dependencies
import pandas as pd
import numpy as np
```

```python
#importing data
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'

data = pd.read_csv(url)

data.head()
```

```python
# Numbr of rows and columns
data.shape
```

```python
# There is one null value in Clearances off line
data.isnull().sum()
```

```python
#What is the index of the NaN value?
data[data['Clearances off line'].isna()]
```

```python
#Will fill NaN value here with 0

df = data.copy()
```

```python
df['Clearances off line'].fillna(0, inplace=True)

df.isna().sum() # all set no NaN values
```

## **1. How many teams participated in the Euro2012?**

```python
df['Team'].value_counts() #Making sure all are unique
```

```python
df['Team'].nunique() #number of unique teams
```

## **Answer:** 16 teams participated in Euro2012.

## **2. What is the number of columns in the dataset?**

```python
df.shape
```

## **Answer:** There are 35 columns in this set.

## **3. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline.**

```python
# A dataframe of the 3 specified columns
discipline = df[['Team', 'Yellow Cards', 'Red Cards']]
```

## **4. Sort the teams by Red Cards, then to Yellow Cards.**

```python
discipline.sort_values(['Red Cards', 'Yellow Cards']) # sorted both ascending by default
```

## **5.Calculate the mean Yellow Cards given per Team.**

```python
df['Yellow Cards'].mean()
```

## **6. Filter teams that scored more than 6 goals.**

```python
df[['Team', 'Goals']][df['Goals']>6]
```

## **Answer:** Germany and Spain scored more than 6 goals

## **7. Select the teams that start with the letter G.**

```python
df['Team'][df['Team'].str[0] == 'G']
```

## **8. Select the first 7 columns.**

```python
df.iloc[:,:7]
```

## **9. Select all columns except the last 3.**

```python
df.iloc[:,:-3]
```

## **10. Present only the Shooting Accuracy from England, Italy and Russia.**

```python
eir = ['England', 'Italy','Russia']
```
